```julia
tbl_files = glob("$(dirn)/tbl*.csv")
order_all_df = DataFrame()
for tbl_file in tbl_files
    order_all_df = vcat(order_all_df, CSV.read(tbl_file, DataFrame))
end
```

## 6

```julia
order_df = @chain order_all_df begin
    @rsubset(:store_id != 999)
end
```

## 7

```julia
order_df = @chain order_df begin
    leftjoin(m_store_df, on=:store_id)
end
```

```julia
ifn = "$(dirn)/m_area.csv"
if !isfile(ifn)
    throw(error("There is no file: $(ifn)"))
end
m_area_df = CSV.read(ifn, DataFrame)
order_df = @chain order_df begin
    leftjoin(m_area_df, on=:area_cd)
end
```

## 8

```julia
function get_status_name(status)
    if ismissing(status)
        return Missing
    elseif status == 0
        return "受付"
    elseif status == 1
        return "お支払済"
    elseif status == 2
        return "お渡し済"
    elseif status == 9
        return "キャンセル"
    end
    return Missing
end
    
order_df = @chain order_df begin
    @rtransform begin
        :takeout_name = :takeout_flag == 0 ? "デリバリー" : "お持ち帰り"
        :status_name = get_status_name(:status)
    end
end
```

## 9

```julia
CSV.write("order_data.csv", order_df)
```
